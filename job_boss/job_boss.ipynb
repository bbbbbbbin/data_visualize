```python
data.tail(5)
```

```python
def area(x):
    try:
        return x.split("·")[1]
    except:
        return None
data['地区'] = data['工作地点'].map(area)
data['是否多薪'] = data['薪资'].map(lambda x: '是' if x[-1] == '薪' else '否')
data['是否实习'] = data['薪资'].map(lambda x : 1 if x[-1] == '天' else 0)
```

```python
data_tmp_x = data['地区'].value_counts().keys().tolist()
data_tmp_y = data['地区'].value_counts().values.tolist()
data_tmp_x.append('未知')
data_tmp_y.append(int(data['地区'].isna().sum()))
init_options = opts.InitOpts(theme='chalk',width='730px')
pie = pyecharts.charts.Pie(init_opts=init_options)
pie = pie.set_global_opts(
    title_opts=opts.TitleOpts('各地区占比'),
    legend_opts=opts.LegendOpts(orient='v',pos_right=0)
)
pie = (pie
    .add('', [list(z) for z in zip(data_tmp_x,data_tmp_y)], radius=['20%', '80%'], rosetype="radius")
    .set_series_opts(label_opts=opts.LabelOpts(formatter='{b} : {d}%'))
)
pie.render_notebook()
```

```python
format_function = """
        function(params){
            var a = params.value/284 * 100;
            a = a.toFixed(2);
            return params.name + ' : ' + a + '%';
        }
"""
```

```python
data_tmp_x = data['学历'].value_counts().keys().tolist()
data_tmp_y = data['学历'].value_counts().values.tolist()
init_options = opts.InitOpts(theme='chalk')
bar = pyecharts.charts.Bar(init_opts=init_options)
bar = bar.set_global_opts(
    title_opts=opts.TitleOpts('需求学历')
)
color_function = """
        function (params) {
            if (params.value > 100) {
                return '#fc97af';
            } else if (params.value > 50 && params.value < 100) {
                return '#72ccff';
            }
            return '#f7f494';
        }
"""
bar = (bar
    .add_xaxis(data_tmp_x)
    .add_yaxis('',data_tmp_y,itemstyle_opts=opts.ItemStyleOpts(color=JsCode(color_function)))
    .set_series_opts(tooltip_opts=opts.TooltipOpts(formatter=JsCode(format_function)))
)
```

```python
data_tmp_x = data['年限'].value_counts().keys().tolist()
data_tmp_y = data['年限'].value_counts().values.tolist()
init_options = opts.InitOpts(theme='chalk')
bar2 = pyecharts.charts.Bar(init_opts=init_options)
bar2 = bar2.set_global_opts(
    title_opts=opts.TitleOpts('年限要求', pos_left='35%'),
    xaxis_opts=opts.AxisOpts(axislabel_opts=opts.LabelOpts(rotate=45))
)
color_function = """
        function (params) {
            if (params.value > 40) {
                return '#fc97af';
            } else if (params.value > 30 && params.value < 40) {
                return '#72ccff';
            }
            return '#f7f494';
        }
"""
bar2 = (bar2
    .add_xaxis(data_tmp_x[:5])
    .add_yaxis('',data_tmp_y,itemstyle_opts=opts.ItemStyleOpts(color=JsCode(color_function)))
    .set_series_opts(tooltip_opts=opts.TooltipOpts(formatter=JsCode(format_function)))
)
```

```python
data_tmp_x = data['公司行业'].value_counts().keys().tolist()
data_tmp_y = data['公司行业'].value_counts().values.tolist()
init_options = opts.InitOpts(theme='chalk')
bar3 = pyecharts.charts.Bar(init_opts=init_options)
bar3 = bar3.set_global_opts(
    title_opts=opts.TitleOpts('公司行业'),          #, pos_right='26%'
    xaxis_opts=opts.AxisOpts(axislabel_opts=opts.LabelOpts(rotate=45))
)
color_function = """
        function (params) {
            if (params.value > 35) {
                return '#fc97af';
            } else if (params.value > 20 && params.value < 40) {
                return '#72ccff';
            }
            return '#f7f494';
        }
"""
bar3 = (bar3
    .add_xaxis(data_tmp_x[:7])
    .add_yaxis('',data_tmp_y,itemstyle_opts=opts.ItemStyleOpts(color=JsCode(color_function)))
    .set_series_opts(tooltip_opts=opts.TooltipOpts(formatter=JsCode(format_function)))
)
bar3.render_notebook()
```

```python
data_tmp = pd.pivot_table(data, index='公司行业', columns='是否多薪',values='Unnamed: 0',aggfunc='count')
data_tmp['总'] = data_tmp.fillna(0)['否'] + data_tmp.fillna(0)['是']
data_tmp = data_tmp.sort_values('总',ascending=False)
data_tmp_x = data_tmp.index.tolist()
data_tmp_y = data_tmp['是'].tolist()
data_tmp_y1 = data_tmp['否'].tolist()
data_tmp_y2 = data_tmp['总'].tolist()

init_options = opts.InitOpts(theme='chalk')
bar4 = pyecharts.charts.Bar(init_opts=init_options)
bar4 = bar4.set_global_opts(
    title_opts=opts.TitleOpts('公司行业及是否多薪', pos_left='60%'),
    xaxis_opts=opts.AxisOpts(axislabel_opts=opts.LabelOpts(rotate=45)),
    legend_opts=opts.LegendOpts(pos_right='8%')
)
bar4 = (bar4
    .add_xaxis(data_tmp_x[:7])
    .add_yaxis('是', data_tmp_y, stack='stack1', category_gap='38%')
    .add_yaxis('否', data_tmp_y1, stack='stack1', category_gap='38%')
    .set_series_opts(label_opts=opts.LabelOpts(position='right'))
)
```

```python
data_tmp = data[data['是否实习'] == 0].copy()
data_tmp['薪资'] = data_tmp['薪资'].map(lambda x: x.split('K')[0])
data_tmp['最低'] = data_tmp['薪资'].map(lambda x: eval(x.split('-')[0])*1000)
data_tmp['最高'] = data_tmp['薪资'].map(lambda x: eval(x.split('-')[1])*1000)
data_tmp['平均'] = data_tmp['薪资'].map(lambda x: (eval(x.split('-')[0]) + eval(x.split('-')[1]))/2*1000)

data_tmp1 = pd.pivot_table(data_tmp, index='年限', values=['最低','最高','平均'], aggfunc='mean')
data_tmp_x = ['1年以内', '1-3年', '3-5年', '5-10年','10年以上']
data_tmp_y = [ round(data_tmp1['最低'][i],2) for i in data_tmp_x ]
data_tmp_y1 = [ round(data_tmp1['平均'][i],2) for i in data_tmp_x ]
data_tmp_y2 = [ round(data_tmp1['最高'][i],2) for i in data_tmp_x ]
```

```python
init_options = opts.InitOpts(theme='chalk')
bar5 = pyecharts.charts.Bar(init_opts=init_options)
bar5 = bar5.set_global_opts(
    title_opts=opts.TitleOpts('年限与平均工资'),
    xaxis_opts=opts.AxisOpts(axislabel_opts=opts.LabelOpts(rotate=45)),
)
bar5 = (bar5
    .add_xaxis(data_tmp_x)
    .add_yaxis('', data_tmp_y,stack='1', category_gap='42%')
    .add_yaxis('', data_tmp_y1,stack='1', category_gap='42%')
    .add_yaxis('', data_tmp_y2,stack='1', category_gap='42%')
    .set_series_opts(label_opts=opts.LabelOpts(position='right'))
)
bar5.render_notebook()
```

```python
def money(data,x):
    data_tmp = pd.pivot_table(data, index=x, values=['最低','最高','平均'], aggfunc='mean')
    data_tmp_x = data_tmp.index.tolist()
    data_tmp_y = [ round(data_tmp['最低'][i],2) for i in data_tmp_x ]
    data_tmp_y1 = [ round(data_tmp['平均'][i],2) for i in data_tmp_x ]
    data_tmp_y2 = [ round(data_tmp['最高'][i],2) for i in data_tmp_x ]
    init_options = opts.InitOpts(theme='chalk')
    bar = pyecharts.charts.Bar(init_opts=init_options)
    bar = bar.set_global_opts(
        title_opts=opts.TitleOpts('{}与薪资'.format(x)),
        xaxis_opts=opts.AxisOpts(axislabel_opts=opts.LabelOpts(rotate=45)),
    )
    bar = (bar
        .add_xaxis(data_tmp_x)
        .add_yaxis('', data_tmp_y,stack='1', category_gap='42%')
        .add_yaxis('', data_tmp_y1,stack='1', category_gap='42%')
        .add_yaxis('', data_tmp_y2,stack='1', category_gap='42%')
        .set_series_opts(label_opts=opts.LabelOpts(position='right'))
    )
    return bar.render_notebook()
```

```python
wordcould = {}
for i in data['标签']:
    for j in eval(i):
        if j in wordcould:
            wordcould[j] += 1
        else:
            wordcould[j] = 1
wcl = []
for i in wordcould:
    if wordcould[i] < 0:
        continue
    else:
        wcl.append((i,wordcould[i]))
```

```python
init_options = opts.InitOpts(theme='chalk')
wc = pyecharts.charts.WordCloud(init_opts=init_options)
wc = wc.set_global_opts(
    title_opts=opts.TitleOpts('行业关键词')
)
wc = (wc
    .add('',wcl, shape='circle')
)
wc.render_notebook()
```
